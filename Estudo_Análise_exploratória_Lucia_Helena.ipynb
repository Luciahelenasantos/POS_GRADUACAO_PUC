## Subplot


```python
# configurando um grupo de subplots em um grid de 1 linha e 2 colunas
# e também o tamanho da figura
fig, ax = plt.subplots(2,2, figsize = (8, 8))

# subplot1: histograma da média de 24 horas de Chuva
ax[0,0].hist(df1['md24hs_Chuva'], bins=7, edgecolor='black')
ax[0,0].set_title("Média 24hs Chuva")

# subplot2: histograma da média de 24 horas de Temperatura
ax[0,1].hist(df1['md24hs_Temp'], bins=7, edgecolor='black')
ax[0,1].set_title("Média 24hs Temperatura")

# subplot3: histograma da média de 24 horas da Velocidade do Vento
ax[1,0].hist(df1['md24hs_Vel_Vento'], bins=7, edgecolor='black')
ax[1,0].set_title("Média 24hs Velocidade do Vento")

# subplot4: histograma da média de 24 horas de SO2
ax[1,1].hist(df1['md24hs_SO2'], bins=7, edgecolor='black');
ax[1,1].set_title("Média 24hs de SO2");
```

## Boxplot

```python
# configurando um grupo de subplots em um grid de 1 linha e 2 colunas
# e também o tamanho da figura
fig, ax = plt.subplots(2,2, figsize = (15, 8))

# subplot1: boxplot do comprimento_sepala
ax[0,0].boxplot(df1['md24hs_Chuva'])
ax[0,0].set_title("Média 24hs Chuva")

# subplot2: boxplot da largura_sepala
ax[0,1].boxplot(df1['md24hs_Temp'])
ax[0,1].set_title("Média 24hs Temperatura")

# subplot3: boxplot do comprimento_petala
ax[1,0].boxplot(df1['md24hs_Vel_Vento'])
ax[1,0].set_title("Média 24hs Velocidade do Vento")

# subplot4: boxplot da largura_petala
ax[1,1].boxplot(df1['md24hs_SO2']);
ax[1,1].set_title("Média 24hs de SO2");
```
